## **Encapsulation**

```python
# Protecting your data/functions from the outside world.
class Bank:
    def __init__(self,acc_no,aadhar_no):
        self.acc_no = acc_no
        self.aadhar_no = aadhar_no
        print('Thank you for your details')
    def withdraw(self):
        print('You can withdraw money here')

    def deposit(self):
        print('You can deposit money here')

    def offers(self):
        print("Find great deals on credit card")
    # This is a private function now. We use "__" to make private function
    def __locker(self):  #This function we don't want anyone to access directly.
        print('You can access jewelleries, money here')

    def Manager(self):  # With this fxn we can get the values(access) our private function: Getter Methods
        self.inp = int(input('Enter your account number: '))
        self.ano = int(input('Enter your aadhar number: '))
        if self.inp==self.acc_no and self.ano == self.aadhar_no:
            print('You can access the locker now')
            self.__locker()


ob = Bank(738838432,365712098989)
ob.deposit()
ob.offers()
ob.__locker()
```

# **Abstraction**

```python
# Showing only necessary information to the user and hiding the unncessary implementation.
# To implement abstraction we create abstract classes and abstract methods.
```

```python
from abc import ABC, abstractmethod
import abc  #abc is a module which contains various functions/classes related to abstraction
# From this module (abc) we have imported ABC class and abstract method to implement abstraction
```

```python
# help(abc)
# Abstract Class contains abstract method
```

```python
class Car(ABC): #Inheriting Abstract Class
    @abstractmethod
    def wheels(self):
        pass

    @abstractmethod
    def door(self):
        pass

    @abstractmethod
    def seats(self):
        pass

    @abstractmethod
    def engine(self):
        pass

    @abstractmethod
    def category(self):
        pass

class Skoda(Car):
    def wheels(self):
        print('Forward Wheel Drive')

    def door(self):
        print('4 Doors')

    def seats(self):
        print('5 Seats')

    def engine(self):
        print('1093 BHP Diesel Engine')

    def category(self):
        print('Sedan')


ob = Skoda()
```

```python
class PaymentProcessor(ABC):
    @abstractmethod
    def pay(self):
        pass


class Debit_CreditCard(PaymentProcessor):
    def pay(self):
        self.pin = int(input('Enter your Pin'))
        self.otp = int(input('Enter OTP'))
        print('Transaction Succesfull')
    def check_offers(self):
        print('Great Offers')

class UPI(PaymentProcessor):
    def pay(self):
        inp = input('Enter your application name: ')
        if inp=='PhonePay':
            self.pin = int(input('Enter your UPI PIN'))
            print('Transaction Succesfull')
        elif inp=='GPay':
            self.upi = int(input('Enter 6 digit upi password'))
            print("Transaction Succesfull")

ob = Debit_CreditCard()
```

```python
# How to pass n number of arguments inside a function where n>=0

def sum(*args): # *parameter_name allows us to pass n numbers of parameters to a fxn.
    sum = 0
    for x in range(len(args)):
        sum += args[x]
    return sum

print(sum())
print(sum(10,20))
print(sum(10,20,30))
print(sum(34,43,23,43,25,645,234,654))
```

```python
class Rent_Vehicle:
    def __init__(self):
        print('Available Cars:')
        self.CarDetails()
        print('Available Bikes: ')
        self.bikeDetails()


    def CarDetails(self):
        self.car_names = ['i20','Innova','Hector','Sonet','Baleno']
        print('Displaying available cars: ')
        for x in self.car_names:
            print("Vehicle Name: ",x)

        print('Price for every car: 1500 per/day')

    def bikeDetails(self):
        self.bike_names = ['MT-15','CB300R','CB350','RE-Hunter','Triumph','Activa']
        print('Displaying available bikes: ')
        for x in self.bike_names:
            print("Vehicle Name: ",x)

        print('Price for every bike: 500 per/day')

class Users(Rent_Vehicle):
    def prices(self):
        print('Base Bike Price: 500')
        print('Base Car Price: 1500')

    def book_vehicle(self):
        name = input('Enter Vehicle Name: ')
        for x in self.car_names:
            if x==name:
                inp = int(input('Enter the number of days: '))
                print('Calculating your Bill:')
                bill = inp*1500
                print("Charges: ",bill)
                insurance = bool(input('Do you want to go for insurance 200 per/day. Enter True/False'))
                if insurance==True:
                    insurance_amount = inp*200
                    print("Final Price: ",bill+insurance_amount)
                confirmation = int(input('Enter 1 to confirm: '))
                if confirmation==1:
                    if inp>5:
                        bill = bill-500

                    print('Booking Confirmed')
                    print('Car Name:',name)
                    print('Amount:',bill)
                else:
                    print('Booking Cancelled')
                    return

        for x in self.bike_names:
            if x==name:
                inp = int(input('Enter the number of days: '))
                bill = inp*500
                print(bill)
                insurance = bool(input('Do you want to go for insurance 200 per/day. Enter True/False'))
                if insurance==True:
                    insurance_amount = inp*200
                    print(bill+insurance_amount)

                confirmation = int(input('Enter 1 to confirm: '))
                if confirmation==1:
                    print('Booking Confirmed')
                    print('Bike Name:',name)
                    if inp>5:
                        print('Amount:',bill-500)
                    print('Amount:',bill)
                else:
                    print('Booking Cancelled')

ob = Users()
```

```python
class Message:
    counter = 0
    def __init__(self,sender,receiver,content):
        self.sender = sender
        self.receiver = receiver
        self.content = content
        self.id = Message.counter
        Message.counter+=1

    def display(self):
        return {
            'id': self.id,
            'sender':self.sender,
            'receiver': self.receiver,
            'content':self.content
        }

class Person:
    def __init__(self,username):
        self.username =username

class User(Person):
    def __init__(self,username):
        super().__init__(username)
        self.inbox = [] # List of messages recieved
        self.sent = [] # List of sent messages.

    def send_message(self,receiver_user,content):
        message = Message(self.username,receiver_user.username,content)
        self.sent.append(message)
        receiver_user.inbox.append(message)
        print('From:',self.username,'Delivered to:',receiver_user.username)

    def show_inbox(self):
        print('Inbox of:',self.username)
        for x in self.inbox:
            print(x.display())
            # From the objects stored in inbox we are going to print all the messages recieved via user

    def show_sent(self):
        print('Senf messages via:',self.username)
        for x in self.sent:
            print(x.display())

    # This will print all the messages sent via user


class ChatSystem: #We will use this class to Register Users
    def __init__(self):
        self.users = {}

    def register_users(self,username):
        if username not in self.users:
            self.users[username] = User(username)
            print('User Registerd Succesfully')
            return self.users[username] #Returing the object of UserClass.
        else:
            print('User Already Exists')


chat_app = ChatSystem()  #Adding users in our backend.
keshav = chat_app.register_users('Keshav')
seema = chat_app.register_users('Seema')
monish = chat_app.register_users('Monish')
anagha = chat_app.register_users('Anagha')
kabir = chat_app.register_users('Kabir')
```

```python
class Message:
    message_id = 1
    def __init__(self, sender, receiver, content):

        self.message = {}
        self.message["id"] = message_id
        self.message["sender"] = sender
        self.message["receiver"] = receiver
        self.message["content"] = content
        message_id += 1

    def display(self):
        print("ID:", self.message["id"])
        print("Sender: ", self.message["sender"])
        print("Receiver:", self.message["receiver"])
        print("Content:", self.message["content"])

class Person:
    def __init__(self, username):
        self.username = username
        self.sent_messages = []
        self.received_messages = []
class User(Person):
    def send_message(self):
        receiver_user = input("To whom would you like to send a message? ")
        content = input("Enter your message: ")
        sender_message = Message(self.username, receiver_user, content)
        print("Message details")
        sender_message.display()
        self.sent_messages.append(sender_message)

    def show_inbox(self):
        print("Your Inbox")
        for i in range(len(self.received_messages)):
            print(i)

    def show_sent(self):
        print("Your sent messages")
        for i in range(len(self.sent_messages)):
            print(i)

class ChatSystem:
    def __init__(self):
        self.users = {}
    def register_user(self):
        username = input("Enter new username: ")

        if username in self.users:
            print("Username already exists!")
            return
        new_user = User(username)
        self.users[username] = new_user
        print('User Registered')

obj = ChatSystem()
obj.register_user()
new_ob = ChatSystem()
new_ob.register_user()
```

**Create a list store 5-10 numbers in it and define a function named as remove_element where you need to pass two parameters (list, target). Write a program to remove the target element from your list. Your function should return the list with elemenyt removed.**

```python
# Test Cases:
# INP: [10,20,30,40,50], target: 30
# OP: [10,20,40,50]
# INP: [40,50,60,80,100], target : 15
# OP: Element Not Found

# 10:10 PM.
def remove_element(li,target):
    new_li = []
    for x in range(len(li)):
        if li[x]!=target:
            new_li.append(li[x])

    if len(new_li) == len(li):
        return 'Element Not Found'
    return new_li

li = [50,60,70,80,211]
op = remove_element(li,80)
print(op)
```

**Create a function to sort a list.**


```python
# Test Case: li = [50,20,30,10,40]
# OP: li = [10,20,30,40,50]

def BubbleSort(arr):
    if len(arr)<=1:
        return arr

    for x in range(len(arr)):
        for j in range(1,len(arr)):
            if arr[j] < arr[j-1]:
                temp = arr[j]
                arr[j] = arr[j-1]
                arr[j-1] = temp

    return arr
```
