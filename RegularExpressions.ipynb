**Regular Expressions are the pattersn used to match and manipulate text**

```python
import re

pattern = r"pattern"  #raw string
data = "information"

result = re.findall(pattern,data)

result
```

**Special Characters**


```python
# Dot (.) -> Matches any character

data = "bat, bet, biat, bot, byte"

print(re.findall(r"b..t",data))
# This . can be replaced via any character
```

```python
# Square Bracket [] -> Character Set

info = "cat, cutt, cut, cpt,  cbt"

print(re.findall(r"c[a,p,u]t",info))
```

```python
# Escape \

text = "Order Bill is 1.5678"
print(re.findall(r"\.",text))
# . -> Any character
# \. -> Means a literal dot
```

```python
# Hyphen - inside []

# Used to define range inside character sets []

# [0-9]
# [a-z]
# [A-Z]

data = "a 3 b 1 8 C d I f, x"

print(re.findall(r"[1-5]",data))
print(re.findall(r"[a-f]",data))
print(re.findall(r"[A-D]",data))
```

```python
# Character Classes

address = "My Pin Code is 3240t01 and my age is 25"

# \d -> Used to fetch digits

print(re.findall(r"\d",address))
print(re.findall(r"\d\d",address))
print(re.findall(r"\d\d\d",address))
print(re.findall(r"\d+",address))
print(re.findall(r"\D+",address))  # No Numbers/Digits
```

```python
# Word Characters \w-> Words

weather = "Today it's 48!! degree"

print(re.findall(r"\w",weather))
print(re.findall(r"\w+",weather))
print(re.findall(r"\W",weather))
```

```python
# Whitespace \s

info = "Name: Kartik Age: 26   "
print(re.findall(r"\s+",info))
print(re.findall(r"\S+",info))

# S D W -> Negations
```

```python
# in the given string count the numbers of spaces we have
spaces = "We are learning Regular Expressions which is also known as Regex."
total_spaces = re.findall(r"\s",spaces)
print(len(total_spaces))
```

```python
# In the given names find all the word fy
words = "magnify classify scatter flying lying"
print(re.findall(r"fy",words))
```

```python
# From the string extract all the digits in group of 2.
string = "abs23, a32b1be33, fd78320r212n3, 3b30dm11"
print(re.findall(r"\d\d",string))
```

**Quantifiers**

```python
# * -> 0 or more Greedy
alphabets = "ab b aaaabc"
print(re.findall(r"a*b",alphabets))
```

```python
# ? -> Yes/No 0/1 To make any characeter optional

words = "color, col5ur, collar"
print(re.findall(r"colou?r",words))
```

```python
print(re.findall(r"col.*r",words))
```

```python
# {n} -> Exact count

numbers = "123 4567 895678"
print(re.findall(r"\d{6}",numbers))
```

```python
numbers = "123 45  67895678 9 8"
# {n,} -> At least n digits
print(re.findall(r"\d{3,}",numbers))
```

```python
# {n,m} -> Between n and m
numbers = "563 83742  732632218 73232  3832  1373  93  932 38"
print(re.findall(r"\d{3,5}",numbers))
```

**Anchors**

```python
# Caret ^  -> From where it's starting
# Dollar $  -> Where it's ending
# String Boundaries

text = "Hello World"

print(re.findall(r"^He",text))
print(re.findall(r"World$",text))
```

```python
# Word Boundary \b

words = "cat scatter cater"
print(re.findall(r"\bcat\b",words))
# Here the first cat has been selected because it is at word boundary
```

```python
words = "cate scatter cater"
print(re.findall(r"\Bcat\B",words))
# Cat has selected from scatter as it's not at word boundaries
```

**Grouping and Clustering**

```python
# Paretheses ()
# They can do two things
# (a) Grouping: Allow you to group parts of pattern together
#  (ab+) =>
# (b) -> Multiple Grouping
```

```python
alphabets = "abab abc ab ababab abababab"
print(re.findall(r"(abab)+",alphabets))
```

```python
# Multiple Groups in Regex
info = "Name: Rahul, Age: 16"
print(re.findall(r"Name: (\w+), Age: (\d+)",info))
```

```python
infos = "Name: Rahul, Age: 16 | Name: Kartik, Age: 25"
print(re.findall(r"Name: (\w+), Age: (\d+)",infos))
```

```python
words ="cat scatter cater matter! better scooter"

# Select all the words that contains tter
print(re.findall(r"\b\w*tter\w*\b",words))
```

**Alteration**

```python
# (|)
animals = "cat dog bat rat"

print(re.findall(r"cat|dog|rat",animals))
```

```python
words = "gray grey great"
print(re.findall(r"(a|e)y",words))
```

```python
files = "report.doc image.png data.csv notes.txt"

# We want to select the doc, png , csv
print(re.findall(r"\w+\.(doc|png|csv)",files))
```

```python
# Non-Capturing Group "?:"

print(re.findall(r"\w+\.(?:doc|png|csv)",files))
```

**Lookarounds**

```python
# Conditions

# (?=) -> Positive Lookahead

words = "apple pie, apple juice, apple watch"
print(re.findall(r"apple(?= pie)",words))
```

```python
## (?!) -> Negative Lookahead

words = "apple pie, apple juice, apple watch"
print(re.findall(r"apple(?! pie)",words))
```

```python
# (?<=) -> Positive Lookbehind
names = "Mr. Kartik, Dr. Haritabh, Mrs. Pooja"

print(re.findall(r"(?<=Dr\. )\w+",names))
```

```python
# (?>=) -> Negative Lookbehind



names = "Mr. Kartik, Dr. Haritabh, Mrs. Pooja"

matches = re.findall(r"(?<!Dr\. )\w+", names)

print(matches)
```

```python
names = "Mr. Kartik, Dr. Haritabh, Mrs. Pooja"

print(re.findall(r"(?:Mr\. | Mrs\. )\w+",names))
```

```python
emails = ["test@example.com", "kartik@domain.org" , "invalid.email@com"]

# ^ -> words
# [\w\.]
# +
pattern = r"^[\w\.]+@[\w\.]+\.\w{2,}$"
for x in emails:
    print(x, bool(re.match(pattern,x)))
```
