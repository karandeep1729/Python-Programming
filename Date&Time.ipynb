Computer stores date and time as numbers.

```python
# What is Epoch in Python Date Time
# An epoch is the starting point of time used via computer to measure time.

# Most Systems(Including Python) they use January 1, 1970 as the epoch.

# Time is then stored as the number of seconds that have passed since that moment.
```

```python
import time
print(time.time())
# This means 1755966682 seconds have been passed since 1 Jan, 1970
```

```python
# Epoch is like time's zero point for computers.
# Python counts second from 1 Jan 1970 (the epoch) to represent data and times.
```
