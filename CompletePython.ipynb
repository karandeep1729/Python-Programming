# **Sets**

**Sets are unique and unordered collection of data elements. Sets are mutable**

```python
basket = {50,'Apple','Banana','Grapes','Avocado','Papaya','Pineapple','Avocado','Grapes',50,45.65}
print(basket)
```

```python
# Let's add some new value to our set.
basket.add('Blue berries')  # we use add function to add some new value to our set.
print(basket)
```

```python
# Let's remove some value:
# To remove value from our set, we have 2 functions:
# remove()
# discard()
```

```python
basket.remove('Carrot')
# If element is not present in our set and if we try to remove it using remove() fxn
# it will give us an error.
```

```python
basket.discard('Carrot')
# If element is not present in our set and if we try to remove it using discard() fxn
# it will not give us an error.
print(basket)
```

```python
basket.discard(45.65)
print(basket)
```

```python
if 'Grapes' in basket:
    print('Grapes exist')
else:
    print("Grapes doesn't exist")
```

```python
basket.remove(50)
print(basket)
```

**Crete a function which will return the multiplication table of a number stored inside a list. Make sure to pass that number as a parameter to the function**

```python
def multiplication_table(num):
    result = []
    for x in range(1,11):
        ans = num*x
        result.append(ans)
    return result
```

```python
output_li = multiplication_table(19)
print(output_li)
```
