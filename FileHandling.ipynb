2. ## ***Reading a File***

```python
bike_info = open('MT15.txt','r')
print(bike_info.read())
```

```python
# Context Manager "with statement" directly opens and closes the file

with open('MT15.txt','r') as f:
    content = f.read()
    print(content) #File will be closed after this
```

```python
# f.read()
content
```

```python
# Reading a specific line
with open('MT15.txt','r') as dummy:
    print(dummy.readline())
```

```python
# If you want to store every line one by one into a list
with open('MT15.txt','r') as dummy:
    lines_list = dummy.readlines()
```

```python
print(lines_list)
```

## 4. ***File Handling Exceptions***

```python
t = open('myfile.txt','r')
```

```python
try:
    with open('myfile.txt','r') as f:
        print(f.read())

except FileNotFoundError:
    print('File Not Found')
```

```python
try:
    with open('New.txt','r'):
        print('File ')

except FileNotFoundError:
    print('File Missing')

else:
    print('File Read Succesfully')

finally:
    print('Finished File Operation')
```
